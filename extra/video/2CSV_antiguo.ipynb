```python
import json
import pandas as pd
### 1. From mmpose to 'dataset/videonly'
# Load JSON data
in_path = '/mnt/b/VIDIMU/extra/videosfullsize_mmpose/S40/S40_A01_T01.json'    # CAMBIAR AL ELEGIR OTRO FICHERO
out_path = '/mnt/b/VIDIMU/extra/videosfullsize_mmpose/resultado.csv'
with open(in_path, 'r') as file:
    data = json.load(file)
# Define joint names based on metadata
keypoint_names = data['meta_info']['keypoint_id2name']
# Prepare data for CSV conversion
frames_data = []
for frame in data['instance_info']:
    frame_id = frame['frame_id']
    keypoints = frame['instances'][0]['keypoints']  # assuming only one instance per frame

    frame_dict = {'frame_id': frame_id}
    for i, (x, y, z) in enumerate(keypoints):
        joint_name = keypoint_names[str(i)]
        frame_dict[f'{joint_name}_x'] = x
        frame_dict[f'{joint_name}_y'] = y
        frame_dict[f'{joint_name}_z'] = z

    frames_data.append(frame_dict)
# Convert to DataFrame
df = pd.DataFrame(frames_data)
import os

# Ensure the directory exists
os.makedirs(os.path.dirname(out_path), exist_ok=True)

# Save to CSV
df.to_csv(out_path, index=False)
print(f"Data saved to {out_path}")
```
