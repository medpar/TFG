# Copies the IMU data in the dataset folder for the mmpose model

```python
import os
import shutil
```

```python
input_base_dir = "/mnt/b/VIDIMU/dataset/videoandimus"
output_base_dir = "/mnt/b/VIDIMU/dataset/dataset_motionagformer/videoandimus"
extensions = (".sto", ".raw", ".mot")
```

```python

def copy_files(input_base_dir, output_base_dir, extensions):
    # Get all directories in the input base directory
    subdirectories = [d for d in os.listdir(input_base_dir) if os.path.isdir(os.path.join(input_base_dir, d))]
    
    for subdir in subdirectories:
        input_dir = os.path.join(input_base_dir, subdir)
        output_dir = os.path.join(output_base_dir, subdir)
        
        if not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        for file in os.listdir(input_dir):
            if file.endswith(extensions):
                src_path = os.path.join(input_dir, file)
                dst_path = os.path.join(output_dir, file)
                shutil.copy2(src_path, dst_path)
                print(f"Copied {src_path} to {dst_path}")

```

```python

copy_files(input_base_dir, output_base_dir, extensions)
```
